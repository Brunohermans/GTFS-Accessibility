```python
import geopandas as gpd
```

```python
allhouses = gpd.read_file('clustersKmeans.shp')
ClusterCentreHouses = gpd.read_file('test-version2.shp')
allhouses
```

```python
ClusterCentreHouses = ClusterCentreHouses.drop('geometry', axis=1)
Allhousesdistances = allhouses.merge(ClusterCentreHouses, left_on = 'cluster', right_on = 'cluster').drop_duplicates('osm_id_x')
```

```python
Allhousesdistances.columns
Allhousesdistances1 = Allhousesdistances[['name_x', 'fclass_x', 'osm_id_x', 'long_x', 'latt_x', 'cluster','C_PTS', 'D_C_PTS', 'C_Super', 'D_C_Super', 'C_DoctorsI', 'D_C_Doctor',
       'C_SchoolsI', 'D_C_School', 'C_SportID_', 'D_C_Sport1', 'Accessibil',
       'geometry']].rename(columns = { 'long_x': 'long', 'latt_x': 'latt', 'osm_id_x': 'osm_id','fclass_x':'fclass'})
Allhousesdistances1.to_file('allhousesclustered.shp')
```
