```python
import pandas as pd
import geopandas as gpd
from simpledbf import Dbf5
dbf = Dbf5(r"C:\Users\bruno\Bureau RMC Dropbox\2. RMC Intern\1. Bureau\3. Stagiairs\Bruno\BHermans\qgis\od-matrix.dbf")
df = dbf.to_dataframe()
df
```

```python
df = df.pivot(index = 'origin_id', columns = 'destinatio', values = 'total_cost')
df
```

```python
df['closest_stop'] = df.idxmin(axis=1)
df
```

```python
Origins = df.index.tolist()
Closeststops = df['closest_stop'].tolist()
Distance = []
for i in range(6014):
    Origin = Origins[i]
    Destination = Closeststops[i]
    Distance.append(df.loc[Origin, Destination])
df1 = pd.DataFrame({'Origin': Origins, 'Destination': Closeststops, 'Distance_to_PT':Distance})
df1
```

```python
df1.to_excel('closeststops.xlsx')
```
