```python
dbf1 = Dbf5(r"C:\Users\bruno\Bureau RMC Dropbox\2. RMC Intern\1. Bureau\3. Stagiairs\Bruno\BHermans\qgis\odmatrixsupermarketmetdestination.dbf")
df1 = dbf1.to_dataframe()
df1
```

```python
df1 = df1.pivot(index='origin_id', columns='destinatio', values = 'total_cost')
df1
```

```python
df1['closest_market'] = df1.idxmin(axis=1)
df1
```

```python
Origins = df1.index.tolist()
Closeststops = df1['closest_market'].tolist()
Distance = []
for i in range(902):
    Origin = Origins[i]
    Destination = Closeststops[i]
    Distance.append(df1.loc[Origin, Destination])
df2 = pd.DataFrame({'Origin': Origins, 'Destination': Closeststops, 'Distance_to_SuperMarket':Distance})
df2
```

```python
merged = df.merge(df2, left_on = 'destinatio', right_on = 'Origin')
merged['total_to_supermarket'] = merged['Distance_to_SuperMarket'] + merged['total_cost']
merged
```

```python
L = merged.origin_id.unique().tolist()
distance = []
for i in L:
    distance.append({'origin': str(i), 'distance_to_supermarket': merged[merged['origin_id'] == i].total_to_supermarket.min(),
                    'supermarketid':
                    merged[(merged['origin_id'] == i) &
                   (merged['total_to_supermarket'] == merged[merged['origin_id'] == i]
                   .total_to_supermarket.min())]['Destination'].to_list()[0]
                    })
distances = pd.DataFrame(distance)
distances
```

```python
gdf = gpd.read_file(r"C:\Users\bruno\Documents\J3Q3-4 BEP\repositories\housesincludingdistancecloseststop.shp")
gdf
```

```python
merged1 = gdf.merge(distances, left_on = 'Destinatio', right_on = 'origin')
merged1['to_closest_supermarket_PT'] = merged1['distance_to_supermarket'] + merged1['Distance_t']
merged1 = merged1.drop(['longdiff', 'Lattdiff', 'Euclidian', 'origin', 'Origin', 'distance_to_supermarket'], axis = 1)
merged1 = merged1.rename(columns = {'Distance_t':'to_closest_PT_stop','Destinatio': 'Closest_stop', 'supermarketid': 'closest_supermarket'})
merged1 = merged1[['name', 'fclass', 'osm_id', 'cluster', 'long', 'latt', 'Closest_stop', 'closest_supermarket', 'to_closest_PT_stop', 'to_closest_supermarket_PT','geometry']]
merged1
```

```python
merged1.to_file('Houses_13-05-2020.shp')
```

```python
X = merged1[['to_closest_PT_stop','to_closest_supermarket_PT']]
kmeans = KMeans(n_clusters=10, random_state=0).fit(X)
a = kmeans.labels_
merged1['cluster_test_accessibility'] = a
merged1
```
